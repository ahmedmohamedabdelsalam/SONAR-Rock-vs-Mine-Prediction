```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
# load dataset
data = pd.read_csv("Copy of sonar data.csv")
data
```

```python
data.columns
```

```python
data.info()
```

# Convert categorical variables to numerical)

```python
data["R"] = data["R"].replace({"M": 0, "R": 1})
```

# separating data

```python
x = data.drop(columns="R", axis=1)
y = data["R"]
```

# traning and test

```python
x_train ,x_test ,y_train ,y_test =train_test_split(x,y,test_size=0.2,random_state=2)
```

# model training

```python
model = LogisticRegression()
```

# model fiting

```python
model.fit(x_train,y_train)
```

# model evalution

```python
x_train_prediction = model.predict(x_train)
accuracy_train= accuracy_score(x_train_prediction,y_train)
```

```python
print("accuracy train =",accuracy_train * 100)
```

# test accuracy

```python
x_test_prediction = model.predict(x_test)
accuracy_test = accuracy_score(x_test_prediction,y_test)
```

```python
print("accuracy test =",accuracy_test * 100)
```
